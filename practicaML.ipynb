A mi parecer estas son las columnas que aportan menos información a la predicción, entre otras podemos encontrar el final de contrato, foto, nombre, logo del club, valoración en cada posición, etc. Esta última me parece que no da valor a la predicción por el hecho de ya tenemos el Overall y Potential, que nos indican la media que tiene el jugador. # Manejo de los NaNs

Un NaN, también conocido como Not a Number es un valor nulo que tenemos que elminar para hacer la predicción. Con el siguiente comando podemos visualizar todas las columnas que contienen valores nulos.

```python
df.columns[df.isna().any()].tolist()
```

Necesitamos eliminar estos NaNs para poder llevar a cabo la predicción. Notese que hay columnas de nuestro dataset que contienen String, ya sea el caso de club por ejemplo. En una sola entrada vamos a eliminar algunos NaNs y a la vez convertiremos estos valores mencionados anteriormente a valores a numéricos. 

```python
df = df.dropna(subset = ['Club'])
df = df.dropna(subset = ['Release Clause'])
df = df.dropna(subset = ['LongShots'])

smv = df.pop("Skill Moves")
wf = df.pop("Weak Foot")
clb = df.pop("Club")
pos = df.pop("Position")
nat = df.pop("Nationality")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(nat, prefix='nat').reset_index(drop=True)], axis=1, sort=False)
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(smv, prefix='Skill Moves').reset_index(drop=True)], axis=1, sort=False)
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(wf, prefix='Weak Foot').reset_index(drop=True)], axis=1, sort=False)
df
```

Ahora que ya tenemos los Strings pasados a enteros haremos una última comprobación para ver si quedas más NaNs

```python
df.columns[df.isna().any()].tolist()
```

Hemos conseguido que la mayoría de valores estén en el formato correcto, pero ahora vamos a ver como podemos tratar los Valores, Salarios, etc. Tenemos que pasar el formato actual a numérico, para ello tendremos que eliminar el símbolo de € y multiplicar K por 1.000 y M por 1.000.000

Para realizar este cambio de formato nos apoyaremos en la siguiente función

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

Aplicamos la función a el Valor, Salario y Clausula de recisión de cada jugador

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
df
```

## Predicción

Ahora que tenemos todas las columnas en el formato correcto, haremos uso de la función df.pop para sacar la columna Value, que será nuestra variable a predecir.

```python
val = df.pop("Value")
df
```

Obtenemos la información de la tabla, ahora todo está en formato numérico y se puede proceder a reralizar la predicción.

La predicción que realizaremos se hará de un 20% y un 30% del total del dataset de entrenamiento. Este porcentaje se puede cambiar en test_size 

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.2, random_state=47)
```

```python
len(X_train)
```

Now we train a LinearRegression model.

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

Finally we get a metric $R^2$ for the regression, we use the implementation from <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html"> sickit-learn</a>.

```python
preds = reg.predict(X_test)
```

Tenemos como valor a predecir 600000, nuestra predicción es de 776446. Como podemos observar nos da un resultado de 0,9945. Es una predicción bastante acertada.
